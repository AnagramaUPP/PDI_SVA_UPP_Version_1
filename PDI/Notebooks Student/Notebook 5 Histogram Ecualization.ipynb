````python
import matplotlib.pyplot as plt

def show_image(image, title=''):

    # Crear figura
    plt.figure(figsize=(6,4))

    # Mostrar imagen
    plt.imshow(image,cmap='gray')

    # Agregar título
    plt.title(title)

    # Quitar ejes
    plt.axis('off')

    # Mostrar resultado
    plt.show()
````

````python
import matplotlib.pyplot as plt

# Cargar imagen
clock_image = plt.imread('../Images/clock_image.png')
clock_rgb=clock_image[:,:,:3]
# Mostrar imagen
show_image(clock_rgb, 'Clock Image')

# Mostrar dimensiones de la imagen original
print('Dimensiones imagen original:', clock_image.shape)
# Mostrar dimensiones de la imagen RGB
print('Dimensiones imagen RGB:', clock_rgb.shape)
# Histograma
plt.figure(figsize=(8,4))

plt.hist(clock_rgb.ravel(),
         bins=256)

plt.title('Histograma de la imagen')
plt.xlabel('Intensidad')
plt.ylabel('Número de píxeles')

plt.show()

# Valor mínimo y máximo
print('Valor mínimo:', clock_rgb.min())
print('Valor máximo:', clock_rgb.max())
````

````python
import matplotlib.pyplot as plt
from skimage import exposure

# CARGAR IMAGEN
chest_xray_image = plt.imread('../Images/chest_xray_image.png')

print('Dimensiones imagen original:', chest_xray_image.shape)

# HISTOGRAMA ORIGINAL

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Imagen original
ax1.imshow(chest_xray_image, cmap='gray')
ax1.set_title('Imagen original')
ax1.axis('off')

# Histograma original
ax2.hist(chest_xray_image.ravel(), bins=256)
ax2.set_title('Histograma original')
ax2.set_xlabel('Intensidad de píxel')
ax2.set_ylabel('Número de píxeles')

plt.show()

# ECUALIZACIÓN DE HISTOGRAMA

xray_image_eq = exposure.equalize_hist(chest_xray_image)

# HISTOGRAMA ECUALIZADO
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Imagen ecualizada
ax1.imshow(xray_image_eq, cmap='gray')
ax1.set_title('Imagen ecualizada')
ax1.axis('off')

# Histograma ecualizado
ax2.hist(xray_image_eq.ravel(), bins=256)
ax2.set_title('Histograma ecualizado')
ax2.set_xlabel('Intensidad de píxel')
ax2.set_ylabel('Número de píxeles')

plt.show()
````

# Imagen aérea

En este ejercicio aplicaremos ecualización de histograma para mejorar el contraste de una imagen aérea. La imagen original presenta bajo contraste, por lo que algunos elementos como carreteras, edificios y zonas urbanas son difíciles de distinguir.

La ecualización de histograma redistribuye las intensidades de los píxeles para aumentar el rango dinámico y resaltar detalles presentes en la imagen.

````python
# Importar librerías
import matplotlib.pyplot as plt
from skimage import exposure
# Cargar imagen aérea
image_aerial = plt.imread('../Images/image_aerial.png')


fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Imagen original
ax1.imshow(image_aerial, cmap='gray')
ax1.set_title('Imagen original')
ax1.axis('off')

# Histograma original
ax2.hist(image_aerial.ravel(), bins=256)

ax2.set_title('Histograma original')
ax2.set_xlabel('Intensidad de píxel')
ax2.set_ylabel('Número de píxeles')

plt.show()
# Aplicar ecualización de histograma
image_eq = exposure.equalize_hist(image_aerial)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Imagen ecualizada
ax1.imshow(image_eq, cmap='gray')
ax1.set_title('Imagen ecualizada')
ax1.axis('off')

# Histograma ecualizado
ax2.hist(image_eq.ravel(), bins=256)

ax2.set_title('Histograma ecualizado')
ax2.set_xlabel('Intensidad de píxel')
ax2.set_ylabel('Número de píxeles')

plt.show()
````

# CLAHE: Ecualización adaptativa limitada por contraste

En este ejercicio utilizaremos la técnica **CLAHE** (*Contrast Limited Adaptive Histogram Equalization*) para mejorar el contraste de una imagen a color.

A diferencia de la ecualización global del histograma, CLAHE trabaja sobre pequeñas regiones de la imagen llamadas vecindarios o mosaicos, permitiendo mejorar detalles locales sin sobre amplificar el ruido.

Esta técnica es ampliamente utilizada en:
- imágenes médicas,
- visión artificial,
- fotografía,
- imágenes satelitales,
- y procesamiento industrial.

````python
# Importar librerías
import matplotlib.pyplot as plt
from skimage import data, exposure
# Cargar imagen
original_image = data.coffee()

# Aplicar ecualización adaptativa
adapthist_eq_image = exposure.equalize_adapthist(
    original_image,
    clip_limit=0.03
)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

# Imagen original
ax1.imshow(original_image)
ax1.set_title('Imagen original')
ax1.axis('off')

# Imagen ecualizada
ax2.imshow(adapthist_eq_image)
ax2.set_title('Imagen ecualizada')
ax2.axis('off')

plt.show()
````
