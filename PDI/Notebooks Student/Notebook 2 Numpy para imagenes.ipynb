# Separación del canal rojo de una imagen

Las imágenes RGB están formadas por 3 canales de color:

- Canal rojo (Red)
- Canal verde (Green)
- Canal azul (Blue)

Para acceder al canal rojo de una imagen, se utiliza:

```python
image[:, :, 0]
```

donde:
- `:` selecciona todos los píxeles
- el último índice `0` corresponde al canal rojo

---

## Código

```python
# Cargar imagen
cat_image = data.cat()

# Obtener canal rojo
red = cat_image[:, :, 0]

# Mostrar canal rojo
plt.figure(figsize=(6,4))
plt.imshow(red, cmap='Reds')

# Título
plt.title('Canal Rojo')

# Quitar ejes
plt.axis('off')

# Mostrar resultado
plt.show()
```

---

## Explicación

La instrucción:

```python
red = cat_image[:, :, 0]
```

extrae únicamente los valores correspondientes al canal rojo de la imagen RGB.

# Separación de los canales RGB de una imagen

Las imágenes RGB están compuestas por tres canales de color:

- Rojo (Red)
- Verde (Green)
- Azul (Blue)

Cada canal puede extraerse individualmente utilizando indexación sobre el arreglo NumPy de la imagen.

---

## Código

```python
from skimage import data
import matplotlib.pyplot as plt

# Cargar imagen
cat_image = data.cat()

# Extraer canales RGB
red = cat_image[:, :, 0]
green = cat_image[:, :, 1]
blue = cat_image[:, :, 2]

# Mostrar dimensiones
print("Shape imagen original:", cat_image.shape)
print("Shape canal rojo:", red.shape)
print("Shape canal verde:", green.shape)
print("Shape canal azul:", blue.shape)

# Crear figura
plt.figure(figsize=(15,5))

# Canal rojo
plt.subplot(1,3,1)
plt.imshow(red, cmap='Reds')
plt.title('Canal Rojo')
plt.axis('off')

# Canal verde
plt.subplot(1,3,2)
plt.imshow(green, cmap='Greens')
plt.title('Canal Verde')
plt.axis('off')

# Canal azul
plt.subplot(1,3,3)
plt.imshow(blue, cmap='Blues')
plt.title('Canal Azul')
plt.axis('off')

# Mostrar resultado
plt.show()
```

---

## Explicación

La imagen original posee 3 dimensiones:

```python
(alto, ancho, 3)
```

donde:
- alto → número de filas
- ancho → número de columnas
- 3 → canales RGB

Al separar cada canal:

```python
red = cat_image[:, :, 0]
```

cada color individual pierde la tercera dimensión y se convierte en una matriz bidimensional:

```python
(alto, ancho)
```

Esto ocurre porque cada canal contiene únicamente intensidades de un solo color.

# Volteamos la imagen con la función `flipud`

```python
import numpy as np

# Volteo vertical de la imagen
cat_image_vertical = np.flipud(cat_image)

# Mostrar imagen
plt.figure(figsize=(6,4))
plt.imshow(cat_image_vertical)

# Título
plt.title('Cat Vertical')

# Quitar ejes
plt.axis('off')

# Mostrar resultado
plt.show()
```

La función:

```python
np.flipud()
```

significa:

- `flip` → invertir
- `ud` → *up-down* (arriba-abajo)

Por lo tanto, esta función realiza un volteo vertical de la imagen.## Volteamos la imagen co la función flipud


# Volteamos la imagen con la función `fliplr`

```python
# Volteo horizontal de la imagen
cat_image_horizontal = np.fliplr(cat_image)

# Mostrar imagen
plt.figure(figsize=(6,4))
plt.imshow(cat_image_horizontal)

# Título
plt.title('Cat Horizontal')

# Quitar ejes
plt.axis('off')

# Mostrar resultado
plt.show()
```

La función:

```python
np.fliplr()
```

significa:

- `flip` → invertir
- `lr` → *left-right* (izquierda-derecha)

Por lo tanto, esta función realiza un volteo horizontal de la imagen.

# Función `show_image()`

En procesamiento de imágenes es común repetir constantemente instrucciones como:

- `plt.figure()`
- `plt.imshow()`
- `plt.title()`
- `plt.axis('off')`
- `plt.show()`

Para evitar escribir los mismos comandos repetidamente, es posible crear una función personalizada llamada `show_image()`.

Esta función permite reutilizar código y mostrar imágenes de una manera más rápida y organizada.

---

## Sintaxis de la función

```python
def show_image(image, title=''):
```

donde:
- `image` corresponde a la imagen que se desea mostrar
- `title` corresponde al título de la imagen

---

## Implementación

```python
import matplotlib.pyplot as plt

def show_image(image, title=''):

    # Crear figura
    plt.figure(figsize=(6,4))

    # Mostrar imagen
    plt.imshow(image)

    # Agregar título
    plt.title(title)

    # Quitar ejes
    plt.axis('off')

    # Mostrar resultado
    plt.show()
```

---

## Ejemplo de uso

```python
show_image(cat_image, 'Cat')
```

Con esta función ya no es necesario escribir todas las instrucciones de visualización cada vez que se desea mostrar una imagen. Es importante señalar que se debe poner siempre al inicio del script o en su defecto generar una función por separar para improtarla (temas avanzados)

# Histogramas de imágenes

Un histograma permite visualizar la distribución de intensidades de los píxeles de una imagen.

En imágenes RGB, es posible analizar individualmente cada canal de color:
- Rojo (Red)
- Verde (Green)
- Azul (Blue)

En este ejemplo se obtiene el histograma del canal rojo de la imagen `cat_2`.

---

## Código

```python
import matplotlib.pyplot as plt

# Obtener canal rojo
red_channel = cat_2[:, :, 0]

# Graficar histograma
plt.figure(figsize=(8,4))
plt.hist(red_channel.ravel(), bins=256)

# Título
plt.title('Histograma del Canal Rojo')

# Etiquetas
plt.xlabel('Intensidad de píxel')
plt.ylabel('Frecuencia')

# Mostrar resultado
plt.show()
```

---

## Explicación

La instrucción:

```python
red_channel = cat_2[:, :, 0]
```

extrae únicamente el canal rojo de la imagen RGB.

Posteriormente:

```python
red_channel.ravel()
```

convierte la matriz bidimensional del canal rojo en un vector unidimensional para poder construir el histograma.

Finalmente:

```python
plt.hist()
```

permite visualizar la distribución de intensidades de los píxeles.

# Separación RGB e histogramas de la imagen `cat_2`

```python
import matplotlib.pyplot as plt

# Separar canales RGB
red_channel = cat_2[:, :, 0]
green_channel = cat_2[:, :, 1]
blue_channel = cat_2[:, :, 2]

# Crear figura
plt.figure(figsize=(15,8))

# Canal rojo
plt.subplot(2,3,1)
plt.imshow(red_channel, cmap='gray')
plt.title('Canal Rojo')
plt.axis('off')

# Canal verde
plt.subplot(2,3,2)
plt.imshow(green_channel, cmap='gray')
plt.title('Canal Verde')
plt.axis('off')

# Canal azul
plt.subplot(2,3,3)
plt.imshow(blue_channel, cmap='gray')
plt.title('Canal Azul')
plt.axis('off')

# Histograma canal rojo
plt.subplot(2,3,4)
plt.hist(red_channel.ravel(), bins=256)
plt.title('Histograma Rojo')
plt.xlabel('Intensidad de píxel')
plt.ylabel('Frecuencia')

# Histograma canal verde
plt.subplot(2,3,5)
plt.hist(green_channel.ravel(), bins=256)
plt.title('Histograma Verde')
plt.xlabel('Intensidad de píxel')
plt.ylabel('Frecuencia')

# Histograma canal azul
plt.subplot(2,3,6)
plt.hist(blue_channel.ravel(), bins=256)
plt.title('Histograma Azul')
plt.xlabel('Intensidad de píxel')
plt.ylabel('Frecuencia')

# Ajustar espacios
plt.tight_layout()

# Mostrar resultado
plt.show()
```
