## Objetivo

Aplicar la operación morfológica de erosión para modificar una imagen binaria y analizar su efecto sobre los objetos presentes en la imagen.

## Descripción

La erosión es una de las operaciones fundamentales de la morfología matemática. Su función consiste en reducir o adelgazar las regiones blancas de una imagen binaria.

Esta operación elimina píxeles de los bordes de los objetos y puede utilizarse para:

- Eliminar ruido de pequeñas dimensiones.
- Separar objetos conectados.
- Reducir el tamaño de regiones binarias.
- Preparar imágenes para procesos posteriores de segmentación.

---

## Conceptos previos

### Imagen binaria

Una imagen binaria únicamente contiene dos posibles valores:

- 0 (negro)
- 1 (blanco)

Las operaciones morfológicas trabajan principalmente sobre este tipo de imágenes.

### Erosión

Durante la erosión, un píxel blanco permanecerá en la imagen únicamente si todos los píxeles cubiertos por el elemento estructurante también pertenecen al objeto.

Como consecuencia:

- Los objetos disminuyen de tamaño.
- Los bordes retroceden.
- Los detalles pequeños pueden desaparecer.

````python
from skimage import io
from skimage import morphology
import matplotlib.pyplot as plt

# Cargar imagen en escala de grises
imagen = io.imread('../Images/R5.png', as_gray=True)

# Convertir a imagen binaria
imagen_binaria = imagen > 0.2

# Aplicar erosión
imagen_erosionada = morphology.erosion(imagen_binaria)

# Mostrar resultados
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(imagen, cmap='gray')
ax[0].set_title('Imagen original')
ax[0].axis('off')

ax[1].imshow(imagen_binaria, cmap='gray')
ax[1].set_title('Imagen binaria')
ax[1].axis('off')

ax[2].imshow(imagen_erosionada, cmap='gray')
ax[2].set_title('Imagen erosionada')
ax[2].axis('off')

plt.tight_layout()
plt.show()
````

Dilatación morfológica

## Objetivo

Aplicar la operación morfológica de dilatación para observar cómo aumentan o se expanden las regiones blancas de una imagen binaria.

## Descripción

La **dilatación** es una operación morfológica que agrega píxeles a los bordes de los objetos blancos en una imagen binaria.

Esta operación puede utilizarse para:

- Engrosar objetos.
- Conectar regiones cercanas.
- Rellenar pequeños espacios.
- Recuperar partes del objeto después de una erosión.

---

La función `morphology.dilation()` permite aplicar dilatación sobre una imagen binaria.

## Preguntas de análisis

1. ¿Qué ocurrió con el grosor de la letra después de aplicar la dilatación?
2. ¿Qué diferencia observas entre la erosión y la dilatación?
3. ¿Qué efecto produce utilizar un elemento estructurante `5x5`?
4. ¿En qué casos sería útil aplicar dilatación?

````python
from skimage import io
from skimage import morphology
import matplotlib.pyplot as plt
import numpy as np

# =====================================================
# Cargar imagen
# =====================================================

imagen = io.imread('../Images/world_image_binary.jpg', as_gray=True)

# =====================================================
# Convertir a binaria
# =====================================================

imagen_binaria = imagen > 0.2

# =====================================================
# Elementos estructurantes
# =====================================================

# Cuadrado 3x3
se_cuadrado_3 = morphology.square(3)

# Cuadrado 5x5
se_cuadrado_5 = morphology.square(5)

# Rombo 5x5 (radio 2)
se_rombo = morphology.diamond(2)

# Cruz 5x5
se_cruz = morphology.star(2)

# =====================================================
# Dilataciones
# =====================================================

dilatacion_cuadrado_3 = morphology.dilation(
    imagen_binaria,
    footprint=se_cuadrado_3
)

dilatacion_cuadrado_5 = morphology.dilation(
    imagen_binaria,
    footprint=se_cuadrado_5
)

dilatacion_rombo = morphology.dilation(
    imagen_binaria,
    footprint=se_rombo
)

dilatacion_cruz = morphology.dilation(
    imagen_binaria,
    footprint=se_cruz
)


# =====================================================
# Comparación de dilataciones
# =====================================================

fig, ax = plt.subplots(1,5, figsize=(20,5))

ax[0].imshow(imagen_binaria, cmap='gray')
ax[0].set_title('Original')

ax[1].imshow(dilatacion_cuadrado_3, cmap='gray')
ax[1].set_title('Cuadrado 3x3')

ax[2].imshow(dilatacion_cuadrado_5, cmap='gray')
ax[2].set_title('Cuadrado 5x5')

ax[3].imshow(dilatacion_rombo, cmap='gray')
ax[3].set_title('Rombo 5x5')

ax[4].imshow(dilatacion_cruz, cmap='gray')
ax[4].set_title('Cruz 5x5')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
````

````python
from skimage import io
from skimage import morphology
import matplotlib.pyplot as plt
import numpy as np

# =====================================================
# Cargar imagen
# =====================================================

imagen = io.imread('../Images/world_image_binary.jpg', as_gray=True)

# =====================================================
# Convertir a binaria
# =====================================================

imagen_binaria = imagen > 0.2

# =====================================================
# Elementos estructurantes
# =====================================================

# Cuadrado 3x3
se_cuadrado_3 = morphology.footprint_rectangle((3,3))

# Cuadrado 5x5
se_cuadrado_5 = morphology.footprint_rectangle((5,5))

# Rombo 5x5
se_rombo = morphology.diamond(2)

# Cruz 5x5
se_cruz = morphology.star(2)

# =====================================================
# Dilataciones
# =====================================================

dilatacion_cuadrado_3 = morphology.dilation(
    imagen_binaria,
    footprint=se_cuadrado_3
)

dilatacion_cuadrado_5 = morphology.dilation(
    imagen_binaria,
    footprint=se_cuadrado_5
)

dilatacion_rombo = morphology.dilation(
    imagen_binaria,
    footprint=se_rombo
)

dilatacion_cruz = morphology.dilation(
    imagen_binaria,
    footprint=se_cruz
)

# =====================================================
# Comparación de dilataciones
# =====================================================

fig, ax = plt.subplots(1,5, figsize=(20,5))

ax[0].imshow(imagen_binaria, cmap='gray')
ax[0].set_title('Original')

ax[1].imshow(dilatacion_cuadrado_3, cmap='gray')
ax[1].set_title('Cuadrado 3x3')

ax[2].imshow(dilatacion_cuadrado_5, cmap='gray')
ax[2].set_title('Cuadrado 5x5')

ax[3].imshow(dilatacion_rombo, cmap='gray')
ax[3].set_title('Rombo 5x5')

ax[4].imshow(dilatacion_cruz, cmap='gray')
ax[4].set_title('Cruz 5x5')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
````
