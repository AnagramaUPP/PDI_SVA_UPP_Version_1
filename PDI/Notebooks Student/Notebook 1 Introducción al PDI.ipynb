## Instrucciones

### Respuestas posibles

- Ambas tienen 3 canales para la representación de color RGB.

- `coffee_image` tiene forma `(303, 384)`, escala de grises. Y `coins_image` `(400, 600, 3)`, RGB-3.

- `coins_image` tiene forma `(303, 384)`, escala de grises. Y `coffee_image` `(400, 600, 3)`, RGB-3.

- Ambas son en escala de grises, con una única dimensión de color.

---

## Explicación

En procesamiento de imágenes, la función `shape()` permite conocer la estructura dimensional de una imagen.

- Si la imagen tiene únicamente `(alto, ancho)`, entonces es una imagen en escala de grises.
- Si la imagen tiene `(alto, ancho, 3)`, entonces posee 3 canales de color RGB (Red, Green, Blue).

---

## Código

```python
from skimage import data

# Cargar imágenes
coffee_image = data.coffee()
coins_image = data.coins()

# Mostrar dimensiones
print("Coffee:", coffee_image.shape)
print("Coins:", coins_image.shape)
```
