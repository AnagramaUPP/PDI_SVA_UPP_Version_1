## Objetivo

Observar el efecto del aliasing durante el proceso de reducción de tamaño de una imagen.

## Descripción

En procesamiento digital de imágenes, el **aliasing** es un efecto que aparece cuando una imagen se reduce de tamaño sin aplicar un filtrado adecuado. Como consecuencia, pueden aparecer bordes irregulares, pérdida de detalles y distorsiones visuales.

En este ejercicio realizarás las siguientes operaciones:

1. Cargar una imagen.
2. Rotarla 90° en sentido horario.
3. Reducir su tamaño a una cuarta parte utilizando anti-aliasing.
4. Reducir su tamaño a una cuarta parte sin utilizar anti-aliasing.
5. Comparar los resultados obtenidos.

---

## Paso 1. Importar las librerías necesarias

Ejecuta la siguiente celda para importar las funciones que se utilizarán durante el ejercicio.

```python
from skimage import io
from skimage.transform import rotate, rescale
import matplotlib.pyplot as plt
```

## Paso 2. Cargar la imagen

Lee la imagen y visualízala.

## Paso 3. Rotar la imagen

La función `rotate()` permite girar una imagen un determinado número de grados.

Un ángulo negativo produce una rotación en sentido horario.

### Actividad

Rota la imagen 90° en sentido horario.

```python
imagen=plt.imread('../Images/kitty2.jpg')
imagen_rotada = rotate(imagen, -90)

plt.figure(figsize=(5,5))
plt.imshow(imagen_rotada)
plt.title("Imagen rotada 90°")
plt.axis("off")
plt.show()
```

## Paso 4. Re-escalamiento con anti-aliasing

Cuando una imagen se reduce de tamaño, es recomendable aplicar un filtro anti-aliasing para disminuir la aparición de artefactos visuales.

### Actividad

Reduce la imagen a una cuarta parte de su tamaño original utilizando anti-aliasing.

```python
imagen_con_aa = rescale(
    imagen_rotada,
    1/4,
    anti_aliasing=True,
    channel_axis=2
)

plt.figure(figsize=(5,5))
plt.imshow(imagen_con_aa)
plt.title("Con anti-aliasing")
plt.axis("off")
plt.show()
```

## Paso 5. Re-escalamiento sin anti-aliasing

Ahora realiza exactamente la misma reducción de tamaño, pero sin aplicar el filtro anti-aliasing.

### Actividad

Observa las diferencias respecto al resultado anterior.

```python
imagen_sin_aa = rescale(
    imagen_rotada,
    1/4,
    anti_aliasing=False,
    channel_axis=2
)

plt.figure(figsize=(5,5))
plt.imshow(imagen_sin_aa)
plt.title("Sin anti-aliasing")
plt.axis("off")
plt.show()
```

## Paso 6. Comparación de resultados

Visualiza ambas imágenes para identificar las diferencias producidas por el aliasing.

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))

ax[0].imshow(imagen_con_aa)
ax[0].set_title("Con anti-aliasing")
ax[0].axis("off")

ax[1].imshow(imagen_sin_aa)
ax[1].set_title("Sin anti-aliasing")
ax[1].axis("off")

plt.show()
```
